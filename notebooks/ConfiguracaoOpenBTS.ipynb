# Configuração `OpenBTS`
Configuração de parâmetros.

## 1. Para iniciar o OpenBTS, digite:

```sqlite3
sudo ./OpenBTS
```

Se você vir a seguinte mensagem de alerta, ignore, tudo deu certo e seu OpenBTS está instalado corretamente:

ALERT 3074381568 15:15:45.5 TRXManager.cpp:595:getFactoryCalibration: READFACTORY
failed with status 1

Em algumas versões do OpenBTS o openBTSCLI já é aberto automaticamente. Se não for, ele deve ser aberto para configurar algumas variáveis. Se você vir a tela abaixo, o openBTSCLI já está aberto.

```sqlite3
Se não, para iniciar o OpenCLI, digite:
```

## 2. Abra outro terminal (crtl+alt+t) e digite:

```sqlite3
sudo ./OpenBTSCLI
```

Você verá:

```sqlite3
OpenBTS>
```

## 3. Para ver as configurações correntes, digite:

```sqlite3
OpenBTS> config
```

Você verá uma lista de configuração de sua BTS GSM.



## 4. Agora verifique se as seguintes configurações estão corretas:

### 4.1. GSM.Radio.Band - Set this to the GSM band appropriate for your hardware. Para visualizar, digite:

```sqlite3
OpenBTS> config GSM.Radio.Band
```

Se não tiver em 900, mude digitando:

```sqlite3
OpenBTS> config GSM.Radio.Band 900
```

### 4.2. GSM.Radio.C0 - This is the ARFCN. Set it to something appropriate for your band. (se não tiver, mude para GSM.Radio.C0 51), digitando:

```sqlite3
OpenBTS> config GSM.Radio.C0
```

Se não tiver em 51, mude digitando:

OpenBTS> config GSM.Radio.C0 51

### 4.3. Control.LUR.OpenRegistration - This tells OpenBTS to not reject your handset just because your registration server (below) isn't responding. Aqui você deve digitar a seguinte configuração:

OpenBTS> config Control.LUR.OpenRegistration .*

### 4.4. GSM.Radio.RxGain - Receiver gain setting in dB. Aqui você deve digitar a seguinte configuração:
O valor ideal é determinado pelo hardware; 47 dB para RAD1 e 0-10 dB para hardware Ettus. Este parâmetro do banco de dados é estático, mas o ganho do receptor pode ser modificado em tempo real com o comando CLI "rxgain".

```sqlite3
devconfig GSM.Radio.RxGain 2
```

### 4.5. Control.LUR.OpenRegistration – determina o nome da rede. Digite:

```sqlite3
OpenBTS> config GSM.Identity.ShortName GppComNet
```

### 4.6. Outros commando úteis:

```sqlite3
OpenBTS> help (ver todos os commandos disponíveis)
```

E quiser sair (não saia!), digite (crtl+c) para sair do OpenBTSCLI.
